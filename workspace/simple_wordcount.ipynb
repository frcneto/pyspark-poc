```python
import sys
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("spark://spark-master:7077").appName("SparkByIESB").getOrCreate()
```

```python
rdd = spark.sparkContext.textFile("./data/shakespeare.txt")
```

```python
import re
rdd2 = rdd.flatMap(lambda line: re.sub('\s+', ' ', line).lower().split(" "))
```

```python
def stopword(s):
    if s not in ['in','a','an','the']:
        return s
    else:
        return ''


rdd3 = rdd2.map(lambda word: (stopword(word), 1)).reduceByKey(lambda a,b:a +b)
```

```python
for (k,v) in rdd3.collect():
    print(k," ==> ",v)
```
